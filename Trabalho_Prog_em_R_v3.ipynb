# **Análise dos Momentos de Compra e Venda**

```r
# Instalando e carregando as bibliotecas necessárias
install.packages("quantmod")
library(quantmod)

# Definindo o ticker da Vale do Rio Doce
ticker <- "BBAS3.SA"  # BBAS3.SA é o ticker para Banco do Brasil na Bovespa
ticker1 <- "ITUB4.SA"  # ITUB4.SA é o ticker para o Itaú na Bovespa
ticker2 <- "BBDC4.SA"  # ITUB4.SA é o ticker para o Bradesco na Bovespa
# Definindo o período desejado
start_date <- "2020-01-01"
end_date <- Sys.Date()  # Data atual

# Baixando os dados do Banco do Brasil da Bovespa
getSymbols(ticker, src = "yahoo", from = start_date, to = end_date)
# Baixando os dados do Banco Itaú da Bovespa
getSymbols(ticker1, src = "yahoo", from = start_date, to = end_date)
# Baixando os dados do Banco Bradesco da Bovespa
getSymbols(ticker2, src = "yahoo", from = start_date, to = end_date)

# Verificando a estrutura dos dados baixados
str(BBAS3.SA)
str(ITUB4.SA)
str(BBDC4.SA)

# Extraindo os preços ajustados
precos_ajustados <- Cl(BBAS3.SA)
precos_ajustados1 <- Cl(ITUB4.SA)
precos_ajustados2 <- Cl(BBDC4.SA)

```

```r
# Calculando médias móveis de curto e longo prazo
BBAS3.SA$SMA20 <- SMA(precos_ajustados, n = 20)
BBAS3.SA$SMA50 <- SMA(precos_ajustados, n = 50)

ITUB4.SA$SMA20 <- SMA(precos_ajustados1, n = 20)
ITUB4.SA$SMA50 <- SMA(precos_ajustados1, n = 50)

BBDC4.SA$SMA20 <- SMA(precos_ajustados2, n = 20)
BBDC4.SA$SMA50 <- SMA(precos_ajustados2, n = 50)
```

```r
# Plotando Gráficos Candlestick para as ações
chartSeries(bancobrasil,
            type="candlesticks",
            main = "Banco do Brasil",
            theme=chartTheme('white'))

chartSeries(itau,
            type="candlesticks",
            main = "Banco do Brasil",
            theme=chartTheme('white'))

chartSeries(bradesco,
            type="candlesticks",
            main = "Banco do Brasil",
            theme=chartTheme('white'))
```

```r
# Calculando os envelopes de Bollinger
n <- 20  # Período para os envelopes de Bollinger
BBAS3.SA$SDB <- BBands(precos_ajustados, n = n, maType = "SMA", sd = 2)$dn
BBAS3.SA$SDT <- BBands(precos_ajustados, n = n, maType = "SMA", sd = 2)$up

ITUB4.SA$SDB <- BBands(precos_ajustados1, n = n, maType = "SMA", sd = 2)$dn
ITUB4.SA$SDT <- BBands(precos_ajustados1, n = n, maType = "SMA", sd = 2)$up

BBDC4.SA$SDB <- BBands(precos_ajustados2, n = n, maType = "SMA", sd = 2)$dn
BBDC4.SA$SDT <- BBands(precos_ajustados2, n = n, maType = "SMA", sd = 2)$up
```

```r
# Plotando os preços ajustados com médias móveis e envelopes de Bollinger do Banco do Brasil
plot(precos_ajustados, type = "l", col = "blue", main = "Preços Ajustados e Indicadores Técnicos",
     ylab = "Preço Ajustado")
lines(BBAS3.SA$SMA20, col = "red")
lines(BBAS3.SA$SMA50, col = "green")
lines(BBAS3.SA$SDB, col = "orange")
lines(BBAS3.SA$SDT, col = "orange")
```

```r
# Plotando os preços ajustados com médias móveis e envelopes de Bollinger do Banco Itau
plot(precos_ajustados1, type = "l", col = "blue", main = "Preços Ajustados e Indicadores Técnicos",
     ylab = "Preço Ajustado")
lines(ITUB4.SA$SMA20, col = "red")
lines(ITUB4.SA$SMA50, col = "green")
lines(ITUB4.SA$SDB, col = "orange")
lines(ITUB4.SA$SDT, col = "orange")
```

```r
# Plotando os preços ajustados com médias móveis e envelopes de Bollinger do Banco Bradesco
plot(precos_ajustados2, type = "l", col = "blue", main = "Preços Ajustados e Indicadores Técnicos",
     ylab = "Preço Ajustado")
lines(BBDC4.SA$SMA20, col = "red")
lines(BBDC4.SA$SMA50, col = "green")
lines(BBDC4.SA$SDB, col = "orange")
lines(BBDC4.SA$SDT, col = "orange")
```

```r
# Adicionando sugestões de compra e venda do Banco do Brasil
# Se o preço cruzar acima da média móvel de curto prazo e estiver abaixo do envelope superior de Bollinger, é um sinal de compra
# Se o preço cruzar abaixo da média móvel de curto prazo e estiver acima do envelope inferior de Bollinger, é um sinal de venda
signals <- rep(NA, length(precos_ajustados))
signals[precos_ajustados > BBAS3.SA$SMA20 & precos_ajustados < BBAS3.SA$SDT] <- 1  # Compra
signals[precos_ajustados < BBAS3.SA$SMA20 & precos_ajustados > BBAS3.SA$SDB] <- -1  # Venda
points(precos_ajustados[signals == 1], col = "green", pch = 18)
points(precos_ajustados[signals == -1], col = "red", pch = 18)
```

```r
# Adicionando sugestões de compra e venda do Banco Itau
# Se o preço cruzar acima da média móvel de curto prazo e estiver abaixo do envelope superior de Bollinger, é um sinal de compra
# Se o preço cruzar abaixo da média móvel de curto prazo e estiver acima do envelope inferior de Bollinger, é um sinal de venda
signals <- rep(NA, length(precos_ajustados1))
signals[precos_ajustados1 > ITUB4.SA$SMA20 & precos_ajustados1 < ITUB4.SA$SDT] <- 1  # Compra
signals[precos_ajustados1 < ITUB4.SA$SMA20 & precos_ajustados1 > ITUB4.SA$SDB] <- -1  # Venda
points(precos_ajustados1[signals == 1], col = "green", pch = 18)
points(precos_ajustados1[signals == -1], col = "red", pch = 18)
```

```r
# Adicionando sugestões de compra e venda do Banco Itau
# Se o preço cruzar acima da média móvel de curto prazo e estiver abaixo do envelope superior de Bollinger, é um sinal de compra
# Se o preço cruzar abaixo da média móvel de curto prazo e estiver acima do envelope inferior de Bollinger, é um sinal de venda
signals <- rep(NA, length(precos_ajustados2))
signals[precos_ajustados2 > BBDC4.SA$SMA20 & precos_ajustados2 < BBDC4.SA$SDT] <- 1  # Compra
signals[precos_ajustados2 < BBDC4.SA$SMA20 & precos_ajustados2 > BBDC4.SA$SDB] <- -1  # Venda
points(precos_ajustados2[signals == 1], col = "green", pch = 18)
points(precos_ajustados2[signals == -1], col = "red", pch = 18)
```
